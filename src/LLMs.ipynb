```python
import llm4series as ls
```

```python
produto = ls.read_file("data/product.csv", index_col="Date")["Revenue"]
pais = ls.read_file("data/country.csv", index_col="Date")["Revenue"]
cliente = ls.read_file("data/customer.csv", index_col="Date")["Revenue"]
```

```python
train_produto, test_produto = produto.split(start="2010-03-15", end="2011-08-31")
train_pais, test_pais = pais.split(start="2009-12-01", end="2011-08-31")
train_cliente, test_cliente = cliente.split(start="2009-12-01", end="2011-08-31")
```

```python
API_KEY = "********"
BASE_URL = "https://integrate.api.nvidia.com/v1"
```

```python
qwen378b = ls.LLM(
  model="qwen/qwen3.5-397b-a17b",
  api_key=API_KEY,
  base_url=BASE_URL
)

mistral_128b = ls.LLM(
  model="mistralai/mistral-medium-3.5-128b",
  api_key=API_KEY,
  base_url=BASE_URL
)
```

```python
prompt_produto = ls.prompt(type="zero_shot", ts=train_produto, tsformat="csv", tstype="numeric", forecast_horizon=len(test_produto), stl=train_produto.stl())
prompt_pais = ls.prompt(type="zero_shot", ts=train_pais, tsformat="csv", tstype="numeric", forecast_horizon=len(test_pais), stl=train_pais.stl())
prompt_cliente = ls.prompt(type="zero_shot", ts=train_cliente, tsformat="csv", tstype="numeric", forecast_horizon=len(test_cliente), stl=train_cliente.stl())
```

```python
def previsao(model, prompt, n_runs=5):
    resultados = []
    for i in range(n_runs):
        p = i + 1
        print(f"[{'■' * p + '□' * (n_runs - p)}] {p}/{n_runs} Modelo: {model.model}")
        resultado = model.predict(prompt, max_tokens=5000)
        resultados.append(resultado)
    return resultados

def avaliacao(resultados, test):
    metricas = []
    for resultado in resultados:
        metricas.append(ls.metrics(test, resultado.prediction))
    return metricas
```

```python
qwen378b_resultado_produto = previsao(qwen378b, prompt_produto)
qwen378b_resultado_pais = previsao(qwen378b, prompt_pais)
qwen378b_resultado_cliente = previsao(qwen378b, prompt_cliente)

mistral_128b_resultado_produto = previsao(mistral_128b, prompt_produto)
mistral_128b_resultado_pais = previsao(mistral_128b, prompt_pais)
mistral_128b_resultado_cliente = previsao(mistral_128b, prompt_cliente)
```

```python
qwen378b_metricas_produto = avaliacao(qwen378b_resultado_produto, test_produto)
qwen378b_metricas_pais = avaliacao(qwen378b_resultado_pais, test_pais)
qwen378b_metricas_cliente = avaliacao(qwen378b_resultado_cliente, test_cliente)

mistral_128b_metricas_produto = avaliacao(mistral_128b_resultado_produto, test_produto)
mistral_128b_metricas_pais = avaliacao(mistral_128b_resultado_pais, test_pais)
mistral_128b_metricas_cliente = avaliacao(mistral_128b_resultado_cliente, test_cliente)
```

```python
def media(metricas):
  mae_media = sum(m["Revenue"]["mae"] for m in metricas) / len(metricas)
  rmse_media = sum(m["Revenue"]["rmse"] for m in metricas) / len(metricas)
  smape_media = sum(m["Revenue"]["smape"] for m in metricas) / len(metricas)
  mae_sem = ls.sem([m["Revenue"]["mae"] for m in metricas])
  rmse_sem = ls.sem([m["Revenue"]["rmse"] for m in metricas])
  smape_sem = ls.sem([m["Revenue"]["smape"] for m in metricas])
  return {
    "mae_media": mae_media,
    "rmse_media": rmse_media,
    "smape_media": smape_media,
    "mae_sem": mae_sem,
    "rmse_sem": rmse_sem,
    "smape_sem": smape_sem
  }
```

```python
qwen378b_media_produto = media(qwen378b_metricas_produto)
qwen378b_media_pais = media(qwen378b_metricas_pais)
qwen378b_media_cliente = media(qwen378b_metricas_cliente)

mistral_128b_media_produto = media(mistral_128b_metricas_produto)
mistral_128b_media_pais = media(mistral_128b_metricas_pais)
mistral_128b_media_cliente = media(mistral_128b_metricas_cliente)
```

```python
import pandas as pd
def resultado(nome_serie, modelo, resultados, metricas):
    return pd.DataFrame([
        {
            "serie": nome_serie,
            "modelo": modelo,
            "periodos_previstos": len(resultado.prediction),
            "valores_previstos": list(resultado.prediction),
            "SMAPE": metrica.loc["smape"].iloc[0],
            "MAE": metrica.loc["mae"].iloc[0],
            "RMSE": metrica.loc["rmse"].iloc[0],
        }
        for resultado, metrica in zip(resultados, metricas)
    ])

qwen378b_resultado = pd.concat([
    resultado(
        "Produto",
        "qwen/qwen3.5-397b-a17b",
        qwen378b_resultado_produto,
        qwen378b_metricas_produto
    ),
    resultado(
        "País",
        "qwen/qwen3.5-397b-a17b",
        qwen378b_resultado_pais,
        qwen378b_metricas_pais
    ),
    resultado(
        "Cliente",
        "qwen/qwen3.5-397b-a17b",
        qwen378b_resultado_cliente,
        qwen378b_metricas_cliente
    )
], ignore_index=True)

mistral_128b_resultado = pd.concat([
    resultado(
        "Produto",
        "mistralai/mistral-medium-3.5-128b",
        mistral_128b_resultado_produto,
        mistral_128b_metricas_produto
    ),
    resultado(
        "País",
        "mistralai/mistral-medium-3.5-128b",
        mistral_128b_resultado_pais,
        mistral_128b_metricas_pais
    ),
    resultado(
        "Cliente",
        "mistralai/mistral-medium-3.5-128b",
        mistral_128b_resultado_cliente,
        mistral_128b_metricas_cliente
    )
], ignore_index=True)
```

```python
resultado = pd.concat([qwen378b_resultado, mistral_128b_resultado], ignore_index=True)
resultado.to_csv("resultados/resultados_LLMs.csv", index=False)
```
