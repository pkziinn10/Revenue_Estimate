```python
import pandas as pd

xls = pd.ExcelFile("data/online_retail_II.xlsx")

df_2009 = pd.read_excel(xls, "Year 2009-2010")
df_2010 = pd.read_excel(xls, "Year 2010-2011")

df = pd.concat([df_2009, df_2010], ignore_index=True)

print(len(df))
```

```python
df.rename(columns={
  "Invoice": "InvoiceNo",
  "Price": "UnitPrice",
  "Customer ID": "CustomerID"
}, inplace=True)
```
